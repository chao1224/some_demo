```python
import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize'] = (20.0, 15.0)


epoch_list = [100]
batch_size_list = [32, 512, 1024, 4096]
batch_size_color = ['r', 'b', 'y', 'g']
```

# Experiment : LeNet + MNIST

## [sample_01, sample_03] refer to the sample indexes, which range from [0, 49]

## Color Specification

| batch size | color |
| :--- | :----- |
| 32 | red |
| 512 | blue |
| 1024 | yellow |
| 4096 | green |

## Mark Specification

| image number | mark |
| :--- | :----- |
| image 01 | solid line |
| image 02 | dashed line |
| image 03 | square |
| avg | triangle |

```python
#################### three sample indexes ##################
sample_01 = 0
sample_02 = 2
sample_03 = 4
sample_idxes = [sample_01, sample_02, sample_03]
#################### three sample indexes ##################

for epoch_idx, epoch in enumerate(epoch_list):
    for batch_idx, batch in enumerate(batch_size_list):
        file_path = './output/lenet/batch_{}_epoch_{}.npz'.format(batch, epoch)
        if os.path.isfile(file_path):
            data = np.load(file_path)

            label = data['training_label_list']
            loss = data['training_loss_list']
            n = label.shape[0]
            plt.semilogy(range(epoch), loss[:, sample_01], '{}-'.format(batch_size_color[batch_idx]),
                         label='data point 01, label {}, batch size {}'.format(label[0, sample_01], batch))
            plt.semilogy(range(epoch), loss[:, sample_02], '{}--'.format(batch_size_color[batch_idx]),
                         label='data point 02, label {}, batch size {}'.format(label[0, sample_02], batch))
            plt.semilogy(range(epoch), loss[:, sample_03], '{}o'.format(batch_size_color[batch_idx]),
                         label='data point 03, label {}, batch size {}'.format(label[0, sample_03], batch))
            plt.semilogy(range(epoch), np.mean(loss[:, sample_idxes], axis=1), '{}^'.format(batch_size_color[batch_idx]),
                         label='data point avg, batch size {}'.format(batch))

plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),fancybox=True, shadow=True, ncol=2)
plt.ylabel('Training Loss')
plt.xlabel('Epoch Number')
plt.title('Training Loss')
plt.show()
plt.clf()

for epoch_idx, epoch in enumerate(epoch_list):
    for batch_idx, batch in enumerate(batch_size_list):
        file_path = './output/lenet/batch_{}_epoch_{}.npz'.format(batch, epoch)
        if os.path.isfile(file_path):
            data = np.load(file_path)

            label = data['training_norm_list']
            loss = data['training_norm_list']
            n = label.shape[0]
            plt.semilogy(range(epoch), loss[:, sample_01], '{}-'.format(batch_size_color[batch_idx]),
                         label='data point 01, label {}, batch size {}'.format(label[0, sample_01], batch))
            plt.semilogy(range(epoch), loss[:, sample_02], '{}--'.format(batch_size_color[batch_idx]),
                         label='data point 02, label {}, batch size {}'.format(label[0, sample_02], batch))
            plt.semilogy(range(epoch), loss[:, sample_03], '{}o'.format(batch_size_color[batch_idx]),
                         label='data point 03, label {}, batch size {}'.format(label[0, sample_03], batch))
            plt.semilogy(range(epoch), np.mean(loss[:, sample_idxes], axis=1), '{}^'.format(batch_size_color[batch_idx]),
                         label='data point avg, batch size {}'.format(batch))

plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),fancybox=True, shadow=True, ncol=2)
plt.ylabel('Training Gradient Norm')
plt.xlabel('Epoch Number')
plt.title('Training Gradient Norm')
plt.show()
```
